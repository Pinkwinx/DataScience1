```python
def get_random_wikipedia_title():
    base_url = "https://en.wikipedia.org"
    random_url = base_url + "/wiki/Special:Random"
    response = requests.get(random_url)
    soup = BeautifulSoup(response.content, 'html.parser')
    article_title = soup.find("h1", {"id": "firstHeading"}).text
    return article_title
```

```python
article_urls = []

for _ in range(10):
    random_article_title = get_random_wikipedia_title()
    article_url = f"https://en.wikipedia.org/wiki/{random_article_title}"
    article_urls.append(article_url)
    print(article_url)
```

```python
num_hyperlinks_per_article = []
for url in article_urls:
    # Send an HTTP GET request and parse the HTML content
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'lxml')
    
    # Find all the hyperlinks within the article content
    hyperlinks = soup.find_all('a')
    
    # Append the number of hyperlinks to the list
    num_hyperlinks_per_article.append(len(hyperlinks))

```

```python
average_num_hyperlinks = np.mean(num_hyperlinks_per_article)
print("Average number of hyperlinks:", average_num_hyperlinks)
```
